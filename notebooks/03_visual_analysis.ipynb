```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("../data/cleaned_olympics_data.csv")
```

```python
top_countries = df['Country'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis')
plt.title("Top 10 Medal-Winning Countries")
plt.xlabel("Country")
plt.ylabel("Total Medals")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=["#66b3ff", "#ff9999"])
plt.title("Gender Distribution of Medals")
plt.axis('equal')
plt.show()
```

```python
medals_by_year = df.groupby('Year')['Medal'].count()

plt.figure(figsize=(10,5))
sns.lineplot(x=medals_by_year.index, y=medals_by_year.values, marker='o')
plt.title("Total Medals Awarded Over Time")
plt.xlabel("Year")
plt.ylabel("Medals")
plt.grid(True)
plt.show()
```

```python
top_sports = df['Sport'].value_counts().head(5)

plt.figure(figsize=(8,5))
sns.barplot(x=top_sports.values, y=top_sports.index, palette='magma')
plt.title("Top 5 Sports by Total Medals")
plt.xlabel("Medals")
plt.ylabel("Sport")
plt.tight_layout()
plt.show()
```
